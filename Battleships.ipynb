```python
import pygame
import sys
import random

# Constants
WIDTH, HEIGHT = 500, 500
ROWS, COLS = 10, 10
CELL_SIZE = WIDTH // COLS
BG_COLOR = (0, 0, 0)  # White
SHIP_COLOR = (0, 0, 255)     # Blue
HIT_COLOR = (255, 0, 0)      # Red
MISS_COLOR = (0, 255, 0)     # Green

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Battleship")

class Game:
    def __init__(self):
        self.grid = [[0] * COLS for _ in range(ROWS)]  # 0 represents empty cell
        self.ships = [(random.randint(0, ROWS-1), random.randint(0, COLS-1)) for _ in range(5)]  # Generate ship coordinates
        self.clicked = set()  # Keep track of clicked cells

    def draw_grid(self):
        for row in range(ROWS):
            for col in range(COLS):
                cell_rect = pygame.Rect(col * CELL_SIZE, row * CELL_SIZE, CELL_SIZE, CELL_SIZE)
                pygame.draw.rect(screen, BG_COLOR, cell_rect)
                if (row, col) in self.clicked:
                    if (row, col) in self.ships:
                        pygame.draw.rect(screen, SHIP_COLOR, cell_rect)
                    else:
                        pygame.draw.rect(screen, MISS_COLOR, cell_rect)

    def handle_click(self, pos):
        row, col = pos[1] // CELL_SIZE, pos[0] // CELL_SIZE
        if (row, col) not in self.clicked:
            self.clicked.add((row, col))
            if (row, col) in self.ships:
                print("Hit!")
            else:
                print("Miss!")

def main():
    game = Game()

    running = True
    while running:
        screen.fill(BG_COLOR)
        game.draw_grid()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.MOUSEBUTTONDOWN:
                if event.button == 1:  # Left click
                    game.handle_click(pygame.mouse.get_pos())

        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()
```
