```python
import tkinter as tk
from tkinter import messagebox
from tkcalendar import DateEntry  # Import DateEntry widget from tkcalendar library

class CalendarApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Calendar App")
        self.events = {}
        self.selected_date = None

        self.create_widgets()

    def create_widgets(self):
        self.calendar_label = tk.Label(self.master, text="Calendar App", font=("Helvetica", 16))
        self.calendar_label.grid(row=0, column=0, columnspan=3, pady=10)

        self.calendar = tk.Text(self.master, height=10, width=30, state=tk.DISABLED)
        self.calendar.grid(row=1, column=0, columnspan=3, padx=10)

        self.add_event_button = tk.Button(self.master, text="Add Event", command=self.add_event)
        self.add_event_button.grid(row=2, column=0, padx=5, pady=10)

        self.view_events_button = tk.Button(self.master, text="View Events", command=self.view_events)
        self.view_events_button.grid(row=2, column=1, padx=5, pady=10)

        self.exit_button = tk.Button(self.master, text="Exit", command=self.master.destroy)
        self.exit_button.grid(row=2, column=2, padx=5, pady=10)

    def add_event(self):
        add_window = tk.Toplevel(self.master)
        add_window.title("Add Event")

        tk.Label(add_window, text="Select Date:").grid(row=0, column=0, pady=5)
        date_picker = DateEntry(add_window, width=12, background='darkblue', foreground='white', borderwidth=2)
        date_picker.grid(row=0, column=1, pady=5)

        tk.Label(add_window, text="Event:").grid(row=1, column=0, pady=5)
        event_entry = tk.Entry(add_window)
        event_entry.grid(row=1, column=1, pady=5)

        add_button = tk.Button(add_window, text="Add", command=lambda: self.save_event(
            date_picker.get_date().strftime('%Y-%m-%d'),
            event_entry.get()
        ))
        add_button.grid(row=2, column=0, columnspan=2, pady=10)

    def save_event(self, date, event):
        if date not in self.events:
            self.events[date] = []
        self.events[date].append(event)
        messagebox.showinfo("Event Added", f"Event added on {date} successfully!")

    def view_events(self):
        view_window = tk.Toplevel(self.master)
        view_window.title("View Events")

        date_picker = DateEntry(view_window, width=12, background='darkblue', foreground='white', borderwidth=2)
        date_picker.grid(row=0, column=0, pady=10)

        view_button = tk.Button(view_window, text="View Events", command=lambda: self.display_events(
            date_picker.get_date().strftime('%Y-%m-%d')
        ))
        view_button.grid(row=0, column=1, pady=10)

    def display_events(self, date):
        view_window = tk.Toplevel(self.master)
        view_window.title(f"Events on {date}")

        if date in self.events:
            events_list = tk.Listbox(view_window, selectmode=tk.SINGLE)
            for event in self.events[date]:
                events_list.insert(tk.END, event)
            events_list.pack(pady=10)
        else:
            tk.Label(view_window, text="No events for selected date.").pack(pady=10)

if __name__ == "__main__":
    root = tk.Tk()
    app = CalendarApp(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
from datetime import datetime

class CalendarApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Calendar App")
        self.events = {}
        self.selected_date = None

        self.create_widgets()

    def create_widgets(self):
        self.calendar_label = tk.Label(self.master, text="Calendar App", font=("Helvetica", 16))
        self.calendar_label.grid(row=0, column=0, columnspan=3, pady=10)

        self.calendar = tk.Text(self.master, height=10, width=30, state=tk.DISABLED)
        self.calendar.grid(row=1, column=0, columnspan=3, padx=10)

        self.add_event_button = tk.Button(self.master, text="Add Event", command=self.add_event)
        self.add_event_button.grid(row=2, column=0, padx=5, pady=10)

        self.view_events_button = tk.Button(self.master, text="View Events", command=self.view_events)
        self.view_events_button.grid(row=2, column=1, padx=5, pady=10)

        self.exit_button = tk.Button(self.master, text="Exit", command=self.master.destroy)
        self.exit_button.grid(row=2, column=2, padx=5, pady=10)

        # Call to initially display any existing events
        self.update_calendar_text()

    def add_event(self):
        add_window = tk.Toplevel(self.master)
        add_window.title("Add Event")

        tk.Label(add_window, text="Select Date:").grid(row=0, column=0, pady=5)
        date_picker = tk.Entry(add_window)
        date_picker.insert(0, datetime.today().strftime('%Y-%m-%d'))
        date_picker.grid(row=0, column=1, pady=5)

        tk.Label(add_window, text="Event:").grid(row=1, column=0, pady=5)
        event_entry = tk.Entry(add_window)
        event_entry.grid(row=1, column=1, pady=5)

        add_button = tk.Button(add_window, text="Add", command=lambda: self.save_event(
            date_picker.get(),
            event_entry.get()
        ))
        add_button.grid(row=2, column=0, columnspan=2, pady=10)

    def save_event(self, date, event):
        if date not in self.events:
            self.events[date] = []

        self.events[date].append(event)
        messagebox.showinfo("Event Added", f"Event added on {date} successfully!")

        # Update the Text widget to display the added event
        self.update_calendar_text()

    def view_events(self):
        view_window = tk.Toplevel(self.master)
        view_window.title("View Events")

        date_picker = tk.Entry(view_window)
        date_picker.grid(row=0, column=0, pady=10)

        view_button = tk.Button(view_window, text="View Events", command=lambda: self.display_events(
            date_picker.get()
        ))
        view_button.grid(row=0, column=1, pady=10)

    def display_events(self, date):
        view_window = tk.Toplevel(self.master)
        view_window.title(f"Events on {date}")

        if date in self.events:
            events_list = tk.Listbox(view_window, selectmode=tk.SINGLE)
            for event in self.events[date]:
                events_list.insert(tk.END, event)
            events_list.pack(pady=10)
        else:
            tk.Label(view_window, text="No events for selected date.").pack(pady=10)

    def update_calendar_text(self):
        self.calendar.config(state=tk.NORMAL)
        self.calendar.delete(1.0, tk.END)  # Clear existing text

        for date, events in self.events.items():
            self.calendar.insert(tk.END, f"{date}:\n")
            for event in events:
                self.calendar.insert(tk.END, f"  - {event}\n")
            self.calendar.insert(tk.END, "\n")

        self.calendar.config(state=tk.DISABLED)


if __name__ == "__main__":
    root = tk.Tk()
    app = CalendarApp(root)
    root.mainloop()
```
